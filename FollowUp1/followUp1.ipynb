```python
import PIL
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
```

First we read the image using Image from PIL library

```python
img=Image.open("lena_gray_512.tif")
```

To avoid open a new window with the image using PIL, we use matplotlib to display it in the notebook.

```python
plt.imshow(np.asarray(img), cmap='grey')
plt.show()
```

Show the information of the image and then we sabe the image as an array of numbers with two decimals

```python
print(img.size, img.mode, img.format)
```

```python
double=np.asarray(img,dtype=np.float64)
```

Then we create an array full of zeros that has the same size of the image

```python
zeros=np.zeros(shape=(512,512))
zeros
```

We find the middel of the image dividing the length of each side of the image by 2

```python
512/2
```

Once we find the middle of the array we can use the conditional "if" and the Pythagoras Theorem to find every value who are inside of the radius and change the 0 for 1 and create the circle

```python
for i in range(zeros.shape[0]):
    for j in range(zeros.shape[1]):
         if ((i-256)**2+(j-256)**2) < (150**2):
             zeros[j,i]= 1

```

```python
plt.imshow(zeros, cmap="gray")
```

Then we multiply the filtter by the image, this can happen because both are arrays

```python
img2=double*zeros
```

```python
plt.imshow(img2, cmap="gray")
```

To display the rest of the image with half intesity we need to change the values that are 0 to 0.5 using a conditional

```python
for i in range(zeros.shape[0]):
    for j in range(zeros.shape[1]):
         if zeros[i,j]==0:
              zeros[i,j]=0.5
```

```python
plt.imshow(zeros)
```

Finally we display the image

```python
img3=zeros*img
plt.imshow(img3, cmap="gray")
```
